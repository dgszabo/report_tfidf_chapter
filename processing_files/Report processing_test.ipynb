```python
#read in text file
with open('Brd A & O Johansen - CSR - 2013.txt') as file:
    testtext = file.read()
```

```python
#tokenizing the text
import nltk
from nltk import word_tokenize
text_tokens = word_tokenize(testtext)
```

```python
#create lowercase
text_tokens_lc = [word.lower() for word in text_tokens]
```

```python
#clean stopwords
from nltk.corpus import stopwords
text_tokens_clean1 = [word for word in text_tokens_lc if word not in stopwords.words('english')]
```

```python
#get rid of punctuations
punctuation = [".", ";", ",", "'", '"', "!", "’", "%", ":", "&", "(", ")", '"', "#", "$", "'",
               "*", "+", "-", "/", "<", "=", ">", "?", "@", "[", "]", "\\", "^", "_", "`", "{",
               "|", "}", "~", '•']

text_tokens_clean2 = [word for word in text_tokens_clean1 if word not in punctuation]
```

```python
#word frequencies
word_frequency = nltk.FreqDist(text_tokens_clean2)

#Print out the most common 5
print(word_frequency.most_common(5))
```

```python
word_frequency.plot(25, cumulative=True)
word_frequency.plot(25, cumulative=False)
```
