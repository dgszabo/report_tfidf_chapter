```python
#import pandas
import pandas as pd

#import data frame
df = pd.read_csv("/Users/dgszabo/Dropbox/Dani/ASB-AU archive/Conferences and Courses/2017-09-Non-financial reporting/Reports/Test selection/Full lists/list_len.csv")

```

```python
#DTM scikit-learn
df['Body'] = df['Body'].apply(lambda x: ''.join([i for i in x if not i.isdigit()]))
```

```python
#import the function CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer

countvec = CountVectorizer()

sklearn_dtm = CountVectorizer().fit_transform(df)
print(sklearn_dtm)
```

```python
#we do the same as we did above, but covert it into a Pandas dataframe. Note this takes quite a bit more memory, so will not be good for bigger data.
dtm_df = pd.DataFrame(countvec.fit_transform(df.Body).toarray(), columns=countvec.get_feature_names(), index = df.index)

#view the dtm dataframe
#dtm_df
```

```python
#most frequent words via DTM
#print(dtm_df.sum().sort_values(ascending=False))
```

```python
##Print the average number of times each word is used in a report
#Print this out sorted from highest to lowest.
print('__________GoldStar 2__________')
print(dtm_df.mean().sort_values(ascending=False)[0:10])

with open('term_frequency_FULL.csv', 'w') as csvfile:
    csv_tf = (dtm_df.mean().sort_values(ascending=False)[0:20000])
    csv_tf.to_csv(csvfile)
```

```python
#import the TFIDF function
from sklearn.feature_extraction.text import TfidfVectorizer

tfidfvec = TfidfVectorizer()

#create the dtm, but with cells weigthed by the tf-idf score.
dtm_tfidf_df = pd.DataFrame(tfidfvec.fit_transform(df.Body).toarray(), columns=tfidfvec.get_feature_names(), index = df.index)

#view results
dtm_tfidf_df
```

```python
print(dtm_tfidf_df.max().sort_values(ascending=False)[0:50])
```
