```python
import glob
import csv
import pandas as pd
from nltk.corpus import stopwords
from nltk import word_tokenize 
list_of_files = glob.glob('/Users/dgszabo/Dropbox/Dani/ASB-AU archive/Conferences and Courses/2017-09-Non-financial reporting/Reports/Test selection/All_txt/*.txt')
dash = "-"
slash = "/"
space = " "
punctuation = [".", ";", ",", "'", '"', "!", "’", "%", ":", "&", "(", ")", '"', "#", "$", "'",
               "*", "+", "-", "/", "<", "=", ">", "?", "@", "[", "]", "\\", "^", "_", "`", "{",
               "|", "}", "~", '•', "...", "–", '“', '”']
years = ["2007", "2008", "2009", "2010", "2011", "2012", "2013", "2014", "2015", "2016", "2017"]
clutter = ["j+", "1", "also", "bang", "olufsen", "dfds", "simcorp", "gabriel", "jyske", 
           "genmab", "per", "'s", "rockwool", "a/s", "\uf0b7", "firstfarms", "royal", "unibrew", 
           "one", "eurosearch", "\x9a\x01", "within", "including", "alk", "arkil", "0", "3",
           "kvistgård", "harboe", "s\x00", "2009/10", "4", "10", "--", "∙∙", "page", "6", "5",
           "2014/15", "zealand", "matas", "h+h", "sydbank", "2013/14", "ion", "vestjyskbank",
           "bavarian", "neurosearch", "2006/2007", "12", "33", "34", "greentech", "rtx", "rias",
           "rias", "lcp", "ambu", "sp", "roblon", "ao", "schouw", "vestjysk", "telecom", "tk",
           "aarsleff", "petroleum", "banknordik", "aircrete", "veloxis", "abelló", "højgaard", 
           "spar", "nord", "atlantic", "pharma", "mr", "mt", "alk", "tat", "søndagsavisen", 
           "aak", "firstfarms", "thorsen", "monberg", "nordic", "and", "phthalate", "brfkredit", 
           "pf", "__________________________", "ges", "fo", "sffoufdi", "abello", "pg", "fo",
           "biomar", "hb", "kirkebjerg"]
data_dic = {}
df = pd.DataFrame(data=None, index=None, columns=["Company","Report type", "Year", "Length", "Body"], dtype=None, copy=False)

row = 0

for file_name in list_of_files:
    fin = file_name.rfind(dash[0])
    fine = file_name.rfind(slash[0])
    
    data_dic["Company"] = file_name[fine+1:fin-7]
    data_dic["Report type"] = file_name[fin-4:fin-1]
    data_dic["Year"] = file_name[fin+2:fin+6]
    
    with open(file_name) as file:
        testtext = file.read()
        text_tokens = word_tokenize(testtext)
        text_tokens_lc = [word.lower() for word in text_tokens]
        text_tokens_clean1 = [word for word in text_tokens_lc if word not in stopwords.words('english')]
        text_tokens_clean2 = [word for word in text_tokens_clean1 if word not in punctuation]
        text_tokens_clean3 = [word for word in text_tokens_clean2 if word not in years]
        text_tokens_clean4 = [word for word in text_tokens_clean3 if word not in clutter]
        data_dic["Body"] = space.join(text_tokens_clean4)
        data_line = data_dic.values()
        data_dic["Length"] = len(text_tokens)
    
    row = row + 1
    for key in data_dic.keys():
        df.loc[row,key] = data_dic[key]


    print (file_name[fine+1:])
    
df
with open('list_len.csv', 'w') as csvfile:
    df.to_csv(csvfile)
```
